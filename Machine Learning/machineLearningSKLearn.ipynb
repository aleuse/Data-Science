## Optimización de Features

### Cómo afectan nuestras features a los modelos de Machine Learning

**Features** son las características o atributos que usamos para la predicción; las variables de entrada.

No siempre es mejor tener más features ya que podrían pasar las siguientes cosas:
- Se le abrirá paso al ruido.
- Features irrelevantes para el modelo y aumentará el costo computacional.
- Si existen features valores faltantes entonces se crearán sesgos muy significativos y perderemos la capacidad de predicción.

Una forma de saber si hemos hecho una buena selección de las features es utilizando el sesgo y la varianza: <br>
![](biasVariance.PNG)

El sesgo es qué tan lejos están los datos del centro de la diana lo que se podría traducir en qué tan acertadas son las predicciones del modelo.

La varianza es qué tan agrupados están los datos lo que se podría traducir a qué tanto se parecen las distintas predicciones del modelo.

Nuestro modelo podría caer en dos escenarios que debemos evitar:
- **Underfitting**: significa que nuestro modelo es demasiado simple, no está captando las features y nuestra variable de salida por lo cual debemos investigar variables con más significado o combinaciones o transformaciones para poder llegar a nuestra variable de salida.
- **Overfitting**: significa que nuestro modelo es demasiado complejo y nuestro algoritmo va a intentar ajustarse a los datos que tenemos pero no se va a comportar bien con los datos del mundo real. Si tenemos overfiting lo mejor es intentar seleccionar los features de una manera más crítica descartando aquellos que no aporten información o combinando algunos para quedarnos con la información que realmente importa.

![](overunder.PNG)

Para solucionar estos problemas podemos hacer lo siguiente:
- Aplicar técnicas de reducción de dimensionalidad como PCA.
- Aplicar la técnica de regulación la cual consiste en penalizar las features que no le estén aportando mucho al modelo.
- Hacer un balanceo, ya sea Oversampling o Undersampling, a conjuntos de datos desbalanceados.

### PCA (Análisis de Componentes Principales)

**Por qué usar PCA**

En ML es normal encontrarnos con problemas donde tengamos una enorme cantidad de features en donde hay relaciones complejas entre ellas y la variable que queremos predecir y aplicando PCA podemos reducir la cantidad de features para quedarnos solo con aquellas que más le aporten al modelo.

**Cuándo utilizar PCA**
- Nuestro conjunto de datos tiene un alto número de features y no estamos seguros si todas son relevantes.
- Hay una alta correlación entre variables o las relaciones entre ellas no son tan sencillas.
- Existe overfitting.
- Hay un alto coste computacional.

**En qué consiste PCA**

PCA consiste básicamente en reducir la complejidad del problema seleccionando solo las variables relevantes y combinándolas en nuevas variables que mantengan la mayor cantidad de varianza de los datos.

![](pca.PNG)

**Pasos para llevar a cabo PCA**
1. Calcular la matriz de covarianza para expresar las relaciones entre nuestras features.
2. Hallar los vectores propios y valores propios de la matriz de covarianza para medir la fuerza y variabilidad de estas relaciones.
3. Ordenar y escoger los vectores propios con mayor variabilidad, esto es, los que aportan más información.

Adicionalmente, si tenemos un PC con bajos recursos podemos usar una variación de PCA llamada IPCA (Incremental PCA)

#### Implementando PCA

**Librerías**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.decomposition import IncrementalPCA
from sklearn.decomposition import KernelPCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

**Cargar los datos**

```python
dfHeart =  pd.read_csv("heart.csv")
dfHeart.head()
```

**Partimos el dataset en variables predictoras y en variable target**

```python
# Variables predictoras o features
XHeart = dfHeart.drop("target", axis = 1)
# Variable target
yHeart = dfHeart.target
```

**Creamos los conjuntos de training y testing**

```python
# Dividimos entre training y testing con un 20% para testing y un 80% para training; shuffle para que distribuya bien
XHeartTrain, XHeartTest, yHeartTrain, yHeartTest = train_test_split(XHeart, yHeart, test_size = 0.2, random_state = 42, shuffle = True)
```

**Escalamos los datos**

```python
# Creamos el escalador
scaler = StandardScaler()

# Escalamos las variables predictoras
XHeartTrainScaled = scaler.fit_transform(XHeartTrain) 
XHeartTestScaled = scaler.fit_transform(XHeartTest)
```

**Aplicamos PCA**

```python
# Creamos el modelo para PCA
pca = PCA()
# Entrenamos el modelo y transformamos los datos
XHeartTrainScaled = pca.fit_transform(XHeartTrainScaled)
XHeartTestScaled = pca.fit_transform(XHeartTestScaled)
pca.explained_variance_ratio_
```

Creamos una gráfica para una mejor visualización

```python
# Varianza acumulada
accVar = np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)
# Lista para guardar el número del componente
components = np.arange(0, 35)

plt.figure(figsize = (10, 10))
# Iteramos a través de los componentes y su respectiva varianza acumulada
for component, aV in zip(components, accVar):
    plt.annotate(round(aV, 2), xy = (component, aV - 2.5))
    
plt.plot(accVar, color = "red", marker = "X", linestyle = "--")
plt.xlabel("Componente")
plt.ylabel("Varianza acumulada")
plt.plot()
```

Nos quedaremos con los componentes del 0 al 3.

```python
# Creamos un df con los datos escalados y con el PCA aplicado
XHeartTrainScaledPCA = pd.DataFrame(XHeartTrainScaled)
XHeartTestScaledPCA = pd.DataFrame(XHeartTestScaled)
# Eliminamos los componentes 4 en adelante
XHeartTrainScaledPCA.drop(np.arange(4, 13), axis = 1, inplace = True)
XHeartTestScaledPCA.drop(np.arange(4, 13), axis = 1, inplace = True)
```

**Aplicamos IPCA**

```python
# Creamos el modelo para IPCA; IPCA no manda todos los datos a entrenar al mismo tiempo sino que lo hace por bloques o Batchs
ipca = IncrementalPCA(batch_size = 10)
# Entrenamos el modelo y transformamos los datos
XHeartTrainScaledIPCA = ipca.fit_transform(XHeartTrainScaled)
XHeartTestScaledIPCA = ipca.fit_transform(XHeartTestScaled)
ipca.explained_variance_ratio_
```

Creamos una gráfica para una mejor visualización

```python
# Varianza acumulada
accVar = np.cumsum(np.round(ipca.explained_variance_ratio_, decimals = 4) * 100)
# Lista para guardar el número del componente
components = np.arange(0, 35)

plt.figure(figsize = (10, 10))
# Iteramos a través de los componentes y su respectiva varianza acumulada
for component, aV in zip(components, accVar):
    plt.annotate(round(aV, 2), xy = (component, aV - 2.5))
    
plt.plot(accVar, color = "red", marker = "X", linestyle = "--")
plt.xlabel("Componente")
plt.ylabel("Varianza acumulada")
plt.plot()
```

Nos quedaremos con los componentes del 0 al 3.

```python
# Creamos un df con los datos escalados y con el PCA aplicado
XHeartTrainScaledIPCA = pd.DataFrame(XHeartTrainScaled)
XHeartTestScaledIPCA = pd.DataFrame(XHeartTestScaled)
# Eliminamos los componentes 4 en adelante
XHeartTrainScaledIPCA.drop(np.arange(4, 10), axis = 1, inplace = True)
XHeartTestScaledIPCA.drop(np.arange(4, 10), axis = 1, inplace = True)
```

**Creamos la regresión logística con PCA**

```python
# Creamos el modelo
logisticRegPCA = LogisticRegression(solver = "lbfgs", max_iter = 10000)
# Entrenamos el modelo
logisticRegPCA.fit(XHeartTrainScaledPCA, yHeartTrain)
# Hacemos unas predicciones
predictionTest = logisticRegPCA.predict(XHeartTestScaledPCA)
metrics.accuracy_score(yHeartTest, predictionTest)
```

**Creamos la regresión logística con IPCA**

```python
# Creamos el modelo
logisticRegIPCA = LogisticRegression(solver = "lbfgs", max_iter = 10000)
# Entrenamos el modelo
logisticRegIPCA.fit(XHeartTrainScaledIPCA, yHeartTrain)
# Hacemos unas predicciones
predictionTest = logisticRegIPCA.predict(XHeartTestScaledIPCA)
metrics.accuracy_score(yHeartTest, predictionTest)
```

### Regularización

La regularización consiste en disminuir la complejidad del modelo a través de una penalización aplicada a las variables menos relevantes. Se introduce un poco de sesgo para reducir la varianza de los datos.

![](regularizacion.PNG)

**Pérdida**

El concepto de pérdida hace referencia a qué tan alejadas están nuestras predicciones de los datos reales, o sea, entre menor sea la pérdida mejor será el modelo.

![](perdida.PNG)

La pérdida es la medida que utilizaremos para aplicar la regularización.

**Tipos de regularización**
- L1 Lasso: reduce la complejidad a través de la eliminación de features que no aportan demasiado al modelo. Penaliza las features que aportan poca información volviéndolas cero, eliminando el ruido que producen en el modelo. <br> ![](lasso.PNG)
- L2 Ridge: reduce la complejidad disminuyendo el impacto de ciertas features al modelo. Penaliza las features poco relevantes pero no las vuelve cero sino que limita la información que aportan al modelo. <br> ![](ridge.PNG)
- ElasticNet: es una combinación de las dos anteriores.

**Lasso Vs Ridge**
- No hay un ganador definitivo.
- Si tenemos pocas features que se relacionan directamente con la variable target entonces podríamos probar Lasso.
- Si tenemos varias features que se relacionan  directamente con la variable target entonces podríamos probar Ridge.

#### Implementando regularización con Lasso y Ridge

**Librerías**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

**Carga de los datos**

```python
dfHappy = pd.read_csv("felicidad.csv")
dfHappy.head()
```

**Partimos el dataset en variables predictoras y en variable target**

```python
# Variables predictoras
XHappy = dfHappy.drop(["country", "rank", "score", "high", "low"], axis = 1)
# Variable target
yHappy = dfHappy.score
```

**Creamos los conjuntos de training y testing**

```python
# Dividimos entre training y testing con un 20% para testing y un 80% para training; shuffle para que distribuya bien
XHappyTrain, XHappyTest, yHappyTrain, yHappyTest = train_test_split(XHappy, yHappy, test_size = 0.2, random_state = 42, shuffle = True)
```

**Creamos el modelo de regresión lineal**

```python
# Creamos el modelo
linearReg = LinearRegression()
# Entrenamos el modelo
linearReg.fit(XHappyTrain, yHappyTrain)
# Hacemos unas predicciones
predictionsHappy = linearReg.predict(XHappyTest)
```

**Creamos el modelo Lasso**

```python
# Creamos el modelo; alpha es la penalización o lambda
lasso = Lasso(alpha = 0.2)
# Entrenamos el modelo
lasso.fit(XHappyTrain, yHappyTrain)
# Hacemos unas predicciones
predictionsHappyLasso = lasso.predict(XHappyTest)
```

**Creamos el modelo Ridge**

```python
# Creamos el modelo; alpha es la penalización o lambda
ridge = Ridge(alpha = 1)
# Entrenamos el modelo
ridge.fit(XHappyTrain, yHappyTrain)
# Hacemos unas predicciones
predictionsHappyRidge = ridge.predict(XHappyTest)
```

**Evaluamos la pérdida** 

```python
# Pérdida de la regresión lineal
linearLoss = mean_squared_error(yHappyTest, predictionsHappy)
# Pérdida con Lasso
lassoLoss = mean_squared_error(yHappyTest, predictionsHappyLasso)
# Pérdida con Ridge
ridgeLoss = mean_squared_error(yHappyTest, predictionsHappyRidge)

print(f"Lineal: {linearLoss} \nLasso: {lassoLoss} \nRidge: {ridgeLoss}")
```

**Cómo afectan Lasso y Ridge a los coeficientes**

```python
print(f"Coeficientes Lasso: {lasso.coef_} \n\nCoeficientes Ridge: {ridge.coef_} \n\nCoeficientes modelo Lineal: {linearReg.coef_}")
```

## Regresiones robustas

### Valores atípicos

- Pueden generar sesgos importantes en los modelos de ML.
- A veces no son un problema sino que revelan información sobre la naturaleza de los datos.
- Ayudan en la detección temprana de fallos.

### Regresiones robustas

Son modelos para abordar valores atípicos.

- **RANSAC**: usamos una muestra aleatoria sobre el conjunto de datos y buscamos la muestra que más datos "buenos" logre incluir; se asume que los "malos valores" no tienen valores específicos.
- **Huber Regressor**: no elimina los valores atípicos sino que los penaliza; los datos son tratados como outliers si el error absoluto de la pérdida está por encima de un umbral llamado épsilon (normalmente igual a 1.35).

### Implementando regresión robusta

**Librerías**

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, RANSACRegressor, HuberRegressor
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

**Carga de datos**

```python
dfHappy = pd.read_csv("felicidad.csv")
dfHappy.head()
```

**Partimos el dataset en variables predictoras y en variable target**

```python
# Variables predictoras
XHappy = dfHappy.drop(["country", "rank", "score"], axis = 1)
# Variable target
yHappy = dfHappy.score
```

**Creamos los conjuntos de training y testing**

```python
# Dividimos entre training y testing con un 20% para testing y un 80% para training; shuffle para que distribuya bien
XHappyTrain, XHappyTest, yHappyTrain, yHappyTest = train_test_split(XHappy, yHappy, test_size = 0.2, random_state = 42, shuffle = True)
```

**Crearemos un diccionario con todos los modelos**

```python
# Estimadores o modelos que utilizaremos
estimators = {
    "SVR" : SVR(gamma = "auto", C = 1, epsilon = 0.1),
    "RANSAC" : RANSACRegressor(max_trials = 10000),
    "HUBER" : HuberRegressor(epsilon = 1.35, max_iter = 10000)
}
```

```python
for name, estimator in estimators.items():
    # Entrenar el modelo
    estimator.fit(XHappyTrain, yHappyTrain)
    # Hacer unas predicciones
    predictionsHappy = estimator.predict(XHappyTest)
    # Medimos el MSE
    print(f"{name} MSE: {mean_squared_error(yHappyTest, predictionsHappy)}\n")
```

## Métodos de ensamble

Los métodos de ensamble funcionan muy bien cuando queremos usar distintos estimadores y llegar a una única conclusión. Se basan en combinar diferentes métodos de ML con diferentes configuraciones y aplicar un método para lograr un consenso.

### Bagging

En lugar de obtener la opinión de un solo "experto" consultamos la opinión de varios "expertos" en paralelos e intentamos crear un consenso.

Bagging proviene de Bootstrap Aggragation.
1. Creamos particiones del conjunto de datos aleatorias (uniformes y con reemplazo.
2. Para cada partición se construyen modelos de aprendizaje.
3. La respuesta final es la combinación (por ejemplo, con votos) entre las respuestas individuales.

### Boosting

Le pediremos a un "experto" su opinión sobre un problema, calculamos su posible error y lo usamos para pedirle a otro "experto" su opinión sobre el mismo problema.

El principio detrás de boosting es fortalecer un modelo débil a través de la repetición secuencial.

## Clustering

Los algoritmos de clustering son las estrategias que podemos usar para agrupar los datos de tal manera que todos los datos pertenecientes a un grupo sean lo más similares posible entre sí y lo más diferentes a los otros grupos.

**Casos de aplicación de clustering**
- No conocemos con anterioridad las features de nuestros datos (Aprendizaje no supervisado).
- Queremos descubrir patrones ocultos en los datos.
- Queremos identificar datos atípicos.

**Casos de uso de aplicación**

- Cuando sabemos cuántos grupos “k” queremos en nuestro resultado. Por ejemplo, en una empresa de marketing y sabemos que los segmentos de clientes es bajo, medio alto, en este caso es recomendable usar M-Means, o bien, Spectral Clustering.
- Cuando queremos que el algoritmo descubra la cantidad de "K" grupos óptima según los datos que tenemos. Por otro lado, si no conocemos cuántos grupos o cuántas features tenemos y solo queremos experimentar, la solución puede ser Meanshift, Clustering Jerárquico o DBSCAN.

## Optimización paramétrica

### Validación de nuestro modelo usando Cross Validation

**Tipos de validación**
- **Hold-Out**: se trata de dividir nuestros datos en entrenamiento y prueba utilizando porcentajes fijos.
- **K-Folds**: se plegan los datos K veces y en dichos pliegues utilizamos distintas partes del dataset como entrenamiento y prueba para intentar cubrir todos los datos de ambos conjuntos.
- **LOOCV**: se dividen los datos en entrenamiento y prueba pero se hace entrenamiento con todos los datos salvo uno y este proceso se repite hasta que todos los datos han sido probados.

**Cuándo utilizar Hold-Out**
- Se requiere un prototipado rápido.
- No se tiene mucho conocimiento en ML.
- No se cuenta con abundante poder de cómputo.

**Cuándo utilizar K-Folds**
- Recomendable en la mayoría de los casos.
- Se cuenta con un equipo suficiente para desarrollar ML.
- Se require la integración con técnicas de optimización paramétrica.
- Se tiene más tiempo para las pruebas.

**Cuándo utilizar LOOCV**
- Se tiene gran poder de cómputo.
- Se cuenta con pocos datos para poder dividir por entrenamiento y prueba.
- Cuando se quiere probar todos los casos posibles.

### Implementación de K-Folds con Cross Validation

**Librerías**

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score, KFold
```

**Carga de datos**

```python
dfHappy = pd.read_csv("felicidad.csv")
dfHappy.head()
```

**Partimos el dataset en variables predictoras y en variable target**

```python
# Variables predictoras
XHappy = dfHappy.drop(["country", "rank", "score"], axis = 1)
# Variable target
yHappy = dfHappy.score
```

**Creamos un modelo de árbol de regresión**

```python
# Creamos el modelo
model = DecisionTreeRegressor()
```

Normalmente utilizaríamos *.fit()* para entrenar el modelo pero como usaremos Cross Validation esta se encargará del proceso.

```python
# Validación rápida; modelo; features; target; puntaje a través del error cuadrático
score = cross_val_score(model, XHappy, yHappy, scoring = "neg_mean_squared_error")
score
```

Cada uno de los elementos es un MSE para cada una de las pruebas que hizo Cross Validation; por defecto hizo 5 pliegues de los datos y eso se puede configurar utilizando el parámetro *cv*.

```python
# Validación rápida; modelo; features; target; puntaje a través del error cuadrático
score = cross_val_score(model, XHappy, yHappy, cv = 3, scoring = "neg_mean_squared_error")
score
```

Para trabajar con esos puntajes debemos sacar su media y luego el valor absoluto para obtener el error medio cuadrado promedio.

```python
np.abs(np.mean(score))
```

Para ver mejor cómo funciona Cross Validation usemos K-Fold

```python
# Número de plieges; datos organizados aleatoriamente;
kfold = KFold(n_splits = 3, shuffle = True, random_state = 42)
# Recorremos cada conjunto de training y testing que nos arroja el split de nuestro dataset
# Realiza la partición en training y testing 3 veces, en este caso
for train, test in kfold.split(dfHappy):
    print("\n\n")
    print(train, test, sep = "\n")
```

Esto nos devuelve los índices del dataset según la partición. Los elementos están guardados en las variables de test y train cuando iteramos así que solo restaría pasarlas a algún modelo.

### Optimización paramétrica con Cross Validation

Familiarizados con el concepto de Cross Validation podemos utilizar el mismo principio para lograr automatizar un poco la selección de nuestros modelos. Scikit Learn nos ofrece enfoques para automatizar el proceso de optimización paramétrica. Existen 3 enfoques principales, estos son:
- **Optimización manual**: escogemos el modelo que queremos ajustar, identificamos los parámetros que necesitaremos y probamos distintas combinaciones una a una hasta que encontremos la configuración adecuada.
- **Optimización por grilla de parámetros GridSearchCV**: definimos una o varias métricas que queremos optimizar, identificamos os posibles valores que pueden tener los parámetros, creamos un diccionario con los parámetros y sus posibles valores, usamos Cross Validation y entrenamos el modelo.
- **Optimización por búsqueda aleatorizada RandomizedSearchCV**: definimos escalas de valores para cada uno de los parámetros seleccionados y el sistema probará algunas combinaciones de forma aleatoria y mostrará la mejor.

**GridSearchCV vs RandomizedSearchCV**

**GridSearchCV**
- Cuando se quiere realizar un estudio a fondo sobre las implicaciones de los parámetros.
- Se tiene tiempo.
- Se tiene poder de procesamiento.

**RandomizedSearchCV**
- Cuando se quiere explorar posibles optimizaciones.
- Hay poco tiempo.
- Hay poco poder de procesamiento.

#### Implementando

**Liberías**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
```

**Carga de datos**

```python
dfHappy = pd.read_csv("felicidad.csv")
dfHappy.head()
```

**Partimos el dataset en variables predictoras y en variable target**

```python
# Variables predictoras
XHappy = dfHappy.drop(["country", "rank", "score"], axis = 1)
# Variable target
yHappy = dfHappy.score
```

**Creamos un modelo de Random Forest Regressor**

```python
# Creamos el modelo
rfr = RandomForestRegressor()
```

**Creamos el diccionario con los parámetros a optimizar**

```python
#
params = {
    # Número de estimadores o árboles que componen el bosque aleatorio
    "n_estimators" : range(4, 16),
    # Criterios para evaluar el modelo
    "criterion" : ["squared_error", "absolute_error"],
    # Profundidad máxima del árbol
    "max_depth" : range(2, 11)
}
```

**Buscamos los parámetros óptimos con RandomizedSearchCV**

```python
# Modelo; grilla de parámetros; máximo de muestreos o configuraciones sobre los parámetros; 3 plieges; forma para medir el score
randSearch = RandomizedSearchCV(rfr, params, n_iter = 10, cv = 3, scoring = "neg_mean_absolute_error")
# Buscamos los parámetros óptimos
randSearch.fit(XHappy, yHappy)

print(f"Modelo óptimo: {randSearch.best_estimator_} \nParámetros óptimos: {randSearch.best_params_}")
```
