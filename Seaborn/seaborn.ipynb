## Set

Es un método de seaborn que permite configurar el estilo, fuente y color de las gráficas.

```python
sns.barplot(x = ["A", "B", "C"], y = [1, 3, 2])
plt.show()
```

Configurar el estilo, la paleta de colores y el tamaño de la fuente

```python
# Style es el estilo del gráfico, palette es la paleta de colores, con font selecciono 
# una fuente y con font_scale le asigno un tamaño
sns.set(style = "darkgrid", palette = "dark", font = "Verdana", font_scale = 1)
sns.barplot(x = ["A", "B", "C"], y = [1, 3, 2])
plt.show()
```

```python
sns.set(style = 'whitegrid', palette = 'icefire', font =" DejaVu Sans", font_scale = 4)
sns.barplot(x = ['A','B','C'], y = [1,3,2])
plt.show()
```

Con *sns.reset_orig()* reseteamos todo.

## Parámetros más usados

Para que las imágenes salgan más nítidas.

```python
sns.set(rc={"figure.dpi":300})
```

```python
# Dataset de seaborns
tips = sns.load_dataset('tips')
tips.head()
```

Los parámetros principales son *data*, *x* y *y*; en este caso haremos un gráfico para observar la distribución.

```python
sns.reset_orig() # Resetear
sns.displot(data = tips, x = "total_bill")
plt.show()
```

Agregaremos una variable *Y*.

```python
sns.displot(data = tips, x = "total_bill", y = "tip")
plt.show()
```

Seaborn reconoce que el gráfico de frecuencias no es la mejor opción así que automáticamente lo adapta.


Con *hue* podemos agrupar por una variable de interés.

```python
sns.displot(data = tips, x = "total_bill", y = "tip", hue = "sex")
plt.show()
```

Podemos cambiar el tipo de la gráfica con *kind*.

```python
sns.displot(data = tips, x = "total_bill", hue = "sex", kind = "kde")
plt.show()
```

Personalizamos la gráfica cambiando la paleta, la transparencia y quitando la leyenda.

```python
sns.reset_orig()
sns.displot(data = tips, x = "total_bill", hue = "sex", kind = "kde", legend = False, 
            palette = "icefire", alpha = 0.5)
plt.show()
```

## Distribuciones

Distribuciones orientadas a datos numéricos.

### Histogramas

Gráfico para distribuciones y que se **acumulen** con *cumulative*.

```python
sns.reset_orig()
sns.histplot(data = tips, x = "tip", bins = 15, cumulative = True)
plt.show()
```

Puedo especificar con *stat* si quiero que trabaje con la frecuencia, la probabilidad o la densidad.

```python
sns.histplot(data = tips, x = "tip", bins = 15, stat = "frequency")
plt.show()
```

```python
sns.histplot(data = tips, x = "tip", bins = 15, stat = "probability")
plt.show()
```

```python
sns.histplot(data = tips, x = "tip", bins = 15, stat = "density")
plt.show()
```

Con *multiple* indico cómo quiero presentar mis datos, ya sean uno encima del otro, por capas, uno al lado del otro o lleno.

```python
sns.histplot(data = tips, x = "tip", bins = 15, hue = "sex", stat = "density", multiple = "stack")
plt.show()
```

```python
sns.histplot(data = tips, x = "tip", bins = 15, hue = "sex", stat = "density", multiple = "dodge")
plt.show()
```

```python
sns.histplot(data = tips, x = "tip", bins = 15, hue = "sex", stat = "density", multiple = "fill")
plt.show()
```

### KDE

Un KDE es un diagrama de densidad.

```python
sns.kdeplot(data = tips, x = "tip", hue = "sex")
plt.show()
```

Puedo hacer el gráfico acumulado.

```python
sns.kdeplot(data = tips, x = "tip", hue = "sex", cumulative = True)
plt.show()
```

Con área bajo la curva.

```python
sns.kdeplot(data = tips, x = "tip", hue = "sex", shade = True)
plt.show()
```

Y con *bw_adjust* ajusta la gráfica y los parámetros que manejamos en los ejes.

```python
sns.kdeplot(data = tips, x = "tip", hue = "sex", bw_adjust = True)
plt.show()
```

### Displot

Es un gráfico de distribución.

```python
sns.displot(data = tips, x = "tip", hue = "sex", multiple = "stack", stat = "frequency")
plt.show()
```

Lo mejor de displot es que puedo volver un histograma o un KDE con *kind*.

## Categóricos

### Gráfica de cantidad

Contar cuántos registros hay.

```python
sns.countplot(data = tips, x = "day", hue = "sex")
plt.show()
```

### Diagrama de puntos

```python
sns.swarmplot(data = tips, x = "day", y = "total_bill", hue = "sex")
plt.show()
```

Si quiero separar los puntos puedo utilizar *dodge*.

```python
sns.swarmplot(data = tips, x = "day", y = "total_bill", hue = "sex", dodge = True)
plt.show()
```

### Boxplot

```python
sns.boxplot(data = tips, x = "day", y = "total_bill", hue = "sex")
plt.show()
```

Es posible combinar múltiples gráficos.

```python
sns.boxplot(data = tips, x = "day", y = "total_bill", hue = "sex", dodge = True)
sns.swarmplot(data = tips, x = "day", y = "total_bill", hue = "sex", dodge = True, color = "0", marker = "X")
plt.show()
```

### Violinplot

Es parecido al boxplot pero no muestra los cuartiles sino cómo se concentran los datos.

```python
sns.violinplot(data = tips, x = "day", y = "total_bill", hue = "sex")
plt.show()
```

Con *split* podemos combinar las categorías.

```python
sns.violinplot(data = tips, x = "day", y = "total_bill", hue = "sex", split = True)
plt.show()
```

### Catplot

Es muy parecido a displot ya que podemos adaptarlo al tipo de gráfico que necesitemos.

```python
sns.catplot(data = tips, x = "day", y = "total_bill", hue = "sex")
plt.show()
```

Automáticamente define cuál es el mejor gráfico.

Con *kind* podemos especificar el tipo de gráfico que queremos y podemos agregar otra variable con *col*.

```python
sns.catplot(data = tips, x = "day", y = "total_bill", hue = "sex", kind = "box", col = "time")
plt.show()
```

De esta manera trabajamos con 4 variables.

## Relación

Este tipo de gráfico nos ayuda a ver la correlación entre distintas variables numéricas.

### Scatterplot

```python
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "day")
plt.show()
```

Podemos cambiar el estilo y tamaño de los puntos.

```python
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "day", style = "time", size = "size")
# Para que la leyenda no tape el gráfico vamos a moverla
plt.legend(loc = "center", bbox_to_anchor = (1.12, 0.5))
plt.show()
```

Y es posible cambiar el formato de los puntos.

```python
marcadores = {"Lunch" : "D", "Dinner" : "s"}
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "day", style = "time", size = "size", markers = marcadores)
# Para que la leyenda no tape el gráfico vamos a moverla
plt.legend(loc = "center", bbox_to_anchor = (1.12, 0.5))
plt.show()
```

### Lineplot

```python
sns.lineplot(data = tips, x = "total_bill", y = "tip", hue = "day", style = "time", size = "size")
plt.show()
```

### Relplot

Es como displot o catplot y nos permite seleccionar el tipo de dato que queramos usar; además, se adapta automáticamente.

```python
sns.relplot(data = tips, x = "total_bill", y = "tip", hue = "day", style = "time", size = "size")
plt.show()
```

Y de igual forma podemos agregar una columna.

```python
sns.relplot(data = tips, x = "total_bill", y = "tip", hue = "day", size = "size", col = "time")
plt.show()
```

## Pairplot

Este gráfico correlaciona todas las variables numéricas que hay en el dataset.

```python
sns.pairplot(data = tips, hue = "day")
plt.show()
```

Podemos hacer que muestre solo una esquina eliminando así las gráficas repetidas.

```python
sns.pairplot(data = tips, hue = "day", corner = True)
plt.show()
```

## Joinplot

Joinplot nos permite unir dos gráficas en una sola.

```python
sns.jointplot(data = tips, x = "total_bill", y = "tip", hue = "sex")
plt.show()
```

Podemos agregar más argumentos para poder hacer un mejor análisis.

```python
sns.jointplot(data = tips, x = "total_bill", y = "tip", hue = "sex", kind = "hist",
              marginal_ticks = True, # Muestra una pequeña tabla para la gráfica exterior
              # Diccionario donde puedo agregar más parámetros como el número de bins, si quiero que estén una 
              # junto a la otra, etc
              marginal_kws = dict(bins = 25, multiple = "dodge"))
plt.show()
```

## Heatmap

Heatmap es un tipo de gráfico enfocado a una estructura matricial. Heatmap correlaciona todas las variables numéricas del dataset.

### Correlación

Podemos ver la correlación entre variables numéricas con *.corr()*.

```python
tips.corr()
```

### Heatmap

Para utilizar heatmap debemos tener una estructura matricial, por ello es que le pasamos el resultado de *.corr()*

```python
sns.heatmap(tips.corr())
plt.show()
```

Hay mucha personalización para este gráfico; con *annot* podemos mostrar el valor de la correlación.

```python
sns.heatmap(tips.corr(), annot = True)
plt.show()
```

Con *cmap* podemos modificar los colores.

```python
sns.heatmap(tips.corr(), annot = True, cmap = "icefire")
plt.show()
```

Usando *linewidths* cambiamos el espaciado entre las variables.

```python
sns.heatmap(tips.corr(), annot = True, cmap = "icefire", linewidths = 5)
plt.show()
```

Y utilizando *linecolor* se cambia el color de las líneas de espaciado entre las variables.

```python
sns.heatmap(tips.corr(), annot = True, cmap = "icefire", linewidths = 5, linecolor = "black")
plt.show()
```

Con *vmin* y *vmax* especificamos el valor mínimo y el valor máximo respectivamente.

```python
sns.heatmap(tips.corr(), annot = True, cmap = "icefire", linewidths = 5, vmin = 0.5, vmax = 1)
plt.show()
```

Finalmente, con *cbar* podemos definir si queremos la barra o no.

```python
sns.heatmap(tips.corr(), annot = True, cmap = "icefire", linewidths = 5, cbar = False)
plt.show()
```

Si queremos invertir los colores entonces agregamos *_r* al final del valor del color.

```python
sns.heatmap(tips.corr(), annot = True, cmap = "icefire_r", linewidths = 5, vmin = 0.5, vmax = 1)
plt.show()
```
