# Distribución binomial

```python
import numpy as np
from numpy.random import binomial
from scipy.stats import binom
from math import factorial
import matplotlib.pyplot as plt
```

## Secuencias de eventos tipo Bernoulli
$$
P(x, n; p) = {n \choose x} p^x (1-p)^{n-x} = \frac{n!}{x!(n-x)!}p^x (1-p)^{n-x}
$$

```python
# definición de la distribución binomial 
def miBinomial(x, n, p):
  return (factorial(n) / (factorial(x) * (factorial(n - x)))) * pow(p, x) * pow(1 - p, n - x)
```

```python
# Utilizando mi función binomial
print(f"Mi binomial: {miBinomial(2, 3, 0.5)}")

# Utilizando la función binomial de Scipy
dist = binom(3, 0.5)
dist.pmf(2)
```

## Simulaciones de secuencias con generadores aleatorios

Los generadores aleatorios tienen como propósito simular muestras de datos que resultarían de muestreos en la vida real de procesos aleatorios como lanzar una moneda o un dado.

```python
# Vamos a hacer un experimento generando una muestra de conjuntos de lanzamientos de a 3 monedas  
def barplot(numeroSimulacros):
  valores = [0, 1, 2, 3]
  muestra = []
  for i in range(numeroSimulacros):
    muestra.append(binomial(3, 0.5))
  distribucion_simulada = np.unique(muestra, return_counts = True)[1]/len(muestra)
  distribucion_teorica = [binom(3, 0.5).pmf(k) for k in valores]
  plt.bar(valores, distribucion_teorica, label = "teoría", color = "yellow")
  plt.bar(valores, distribucion_simulada, label = "simulación", alpha = 0.5, color = "blue")
  plt.title(f"simulación con {numeroSimulacros} experimentos")
  plt.show()

barplot(20)
barplot(200)
barplot(20000)
```
