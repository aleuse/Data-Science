# Distribución Normal

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm 
```

## Distribución normal teórica


$$P(X) = \frac{1}{\sigma \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{X-\mu}{\sigma} \right)^2 \right]}$$

* $\mu$: media de la distribución
* $\sigma$: desviación estándar de la distribución

```python
# definimos nuestra distribución normal
def normal(x, mu, sigma):
  return 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * pow((x - mu) / sigma, 2))
```

```python
x = np.arange(-4,4,0.1)
y = normal(x, 0.0, 1.0)

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-4,4,0.1)
# Desplazar con la media y "adelgazar" con la desviación estándar
y = normal(x, 1, 0.5)

plt.plot(x, y)
plt.show()
```

Usando Scipy podemos utilizar el objeto *norm* y luego el atributo *pdf* para obtener la fdp

```python
# usando scipy
dist = norm(0, 1)
x = np.arange(-4, 4, 0.1)
y = [dist.pdf(value) for value in x]
plt.plot(x, y)
plt.show()
```

Con *cdf* obtenemos la fda

```python
# calculando la distribución acumulada
dist = norm(0, 1)
x = np.arange(-4,4,0.1)
y = [dist.cdf(value) for value in x]
plt.plot(x, y)
plt.show()
```

## Distribución normal a partir de datos

```python
df = pd.read_excel('s057.xls')
arr = df['Normally Distributed Housefly Wing Lengths'].values[3:]
values, dist = np.unique(arr, return_counts=True)
print(values)
plt.bar(values, dist)
plt.show()
```

## Estimación de la distribución de probabilidad

```python
# estimación de la distribución de probabilidad


#distribución teórica
mu = arr.mean()
sigma = arr.std()
dist = norm(mu, sigma)
x = np.arange(30, 60, 0.1)
y = [dist.pdf(value) for value in x]
plt.plot(x, y)

# distribución empírica
values, dist = np.unique(arr, return_counts=True)
plt.bar(values, dist/len(arr))
plt.show()
```
