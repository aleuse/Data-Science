## Subplot

Subplot permite crear gráficos dentro de un gráfico. Esto lo hace a través de una matriz de gráficos y se puede acceder a ellos a traves de los índices.

Vamos a crear una matriz de gráficos de una fila y dos columnas.

```python
# #Filas, #Columnas, Índice
# Gráfico 1
plt.subplot(1, 2, 1)
plt.plot(x, y, "--", color = "red")

# Gráfico 2
plt.subplot(1, 2, 2)
plt.hist(y, density = True)

plt.show()
```

Puedo añadir varios plots a un mismo gráfico.

```python
# Gráfico 1
plt.subplot(1, 2, 1)
# Plot 1
plt.plot(x, y, "--", color = "red")
# Plot 2
plt.plot(y, x, ":", color = "blue")

# Gráfico 2
plt.subplot(1, 2, 2)
# Plot 1
plt.pie(y)

plt.show()
```

Crear una matriz de gráficos de dos filas y una columna.

```python
# Gráfico 1
plt.subplot(2, 1, 1)
# Plot 1
plt.plot(x, y, "--", color = "red")
# Plot 2
plt.plot(y, x, ":", color = "blue")

# Gráfico 2
plt.subplot(2, 1, 2)
# Plot 1
plt.hist(x, density = True)

plt.show()
```

## Método orientado a objetos

El método orientado a objeto nos permite mayor personalización y más amigable a múltiples diagramas.

Un objeto define una figura, esa figura es un **lienzo** en el cual podemos introducir diferentes gráficas (**axes**) y cada una de ellas posee sus propios ejes (**axis**). La figura representa el todo y dentro de ella configuramos las gráficas con sus distintos ejes.

![metodo-orientado-objetos](metodo-orientado-objetos.png)

Primero creamos nuestro lienzo, la figura.

```python
figura = plt.figure()
```

Una vez tenemos el lienzo entonces procedemos a agregarle una gráfica o *axes*. 

Parámetros de axes:
- Pos. Eje x	
- Pos. Eje y	
- Size graph eje x	
- Size graph eje y

```python
figura = plt.figure()
# [PosEjeX, PosEjeY, SizeGraphX, SizeGraphY]
axes1 = figura.add_axes([0.1, 0.1, 0.5, 0.9])
axes1.plot(x, y, color = "blue")
figura.show()
```

Ahora creamos otra gráfica dentro del gráfico.

```python
figura = plt.figure()
# Añadimos el gráfico 1
# [PosEjeX, PosEjeY, SizeGraphX, SizeGraphY]
axes1 = figura.add_axes([0.1, 0.1, 0.5, 0.9])
# Añadimos el gráfico 2
axes2 = figura.add_axes([0.17, 0.55, 0.4, 0.3])

# Plot 1
axes1.plot(x, y, color = "blue")
# Plot 2
axes2.plot(y, x, color = "red")

# Cambiar el fondo del axes2
axes2.set_facecolor("gray")

figura.show()
```

Como cada objeto es distinto entonces puedo personalizar cada uno como quiera.

## Subplots

Subplots aprovecha las bondades del método orientado a objetos. 

Con subplots se puede trabajar en un arreglo de gráficas a las cuales se accede a través de los índices.

Creamos una gráfica de tipo figura con un gráfico adentro.

```python
figura, axes = plt.subplots()
axes.plot(x, y, color = "blue")
plt.show()
```

Varias gráficas de tipo figura. En este caso vamos a trabajar con una fila y dos columnas.

```python
figura, axes = plt.subplots(nrows = 1, ncols = 2)
```

Podemos acceder a los gráficos de la figura a través del índice.

```python
figura, axes = plt.subplots(nrows = 1, ncols = 2)
# Gráfico de la izquierda
axes[0].plot(x , y, color = "blue")

# Gráfico de la derecha
axes[1].plot(y, x, color = "red")

plt.show()
```

Podemos acceder a los gráficos mediante el nombre si los tenemos almacenados en una tupla.

```python
figura, (axes1, axes2) = plt.subplots(nrows = 1, ncols = 2)
# Gráfico 1
axes1.plot(x , y, color = "blue")

# Gráfico 2
axes2.plot(y, x, color = "red")

plt.show()
```

Generar una matriz de dos filas y cuatro columnas.

```python
figura, axes = plt.subplots(2, 4)

# Gráfica fila 1, columna 0
axes[1, 0].plot(x, np.cos(x), color = "green")

# Gráfica fila 1, columna 1
axes[1, 1].plot(x, np.sin(x), color = "red")

# Gráfica fila 1, columna 2
axes[1, 2].plot(x, np.tan(x), color = "magenta")

# Gráfica fila 1, columna 3
axes[1, 3].plot(x, np.arctan(x), color = "blue")

# Mejora la visualización de los ejes de cada gráfico
figura.tight_layout()
```

De forma alterna podemos acceder a cada gráfico mediante sus definidos nombres en una tupla.

```python
figura, ( (axes0, axes1, axes2, axes3), (axes4, axes5, axes6, axes7) ) = plt.subplots(2, 4)

# Gráfica fila 0, columna 0
axes0.plot(x, np.cos(x), color = "green")

# Gráfica fila 0, columna 1
axes1.plot(x, np.sin(x), color = "red")

# Gráfica fila 0, columna 2
axes2.plot(x, np.tan(x), color = "magenta")

# Gráfica fila 0, columna 3
axes3.plot(x, np.arctan(x), color = "blue")

# Mejora la visualización de los ejes de cada gráfico
figura.tight_layout()
```

## Leyendas, etiquetas, títulos, tamaño

Para dar contexto a nuestros gráficos necesitamos usar títulos, leyendas, tamaño o etiquetas, para que nuestra gráfica tenga un contexto más amplio.

```python
figura, axes = plt.subplots(1, 2)
axes[0].plot(x, y, color = "blue")
axes[1].plot(y, x, color = "red")
figura.tight_layout()
```

### Títulos

```python
figura, axes = plt.subplots(1, 2)
axes[0].plot(x, y, color = "blue")
# Creamos un título para el gráfico de la izquierda
axes[0].set_title("Relación X - Y")

axes[1].plot(y, x, color = "red")
# Creamos un título para el gráfico de la derecha
axes[1].set_title("Relación Y - X")

figura.tight_layout()
```

### Etiquetas de los ejes

```python
figura, axes = plt.subplots(1, 2)
axes[0].plot(x, y, color = "blue")
# Creamos un título para el gráfico de la izquierda
axes[0].set_title("Relación X - Y")
# Definimos la etiqueta del eje X de la gráfica de la izquierda
axes[0].set_xlabel("Eje X")

axes[1].plot(y, x, color = "red")
# Creamos un título para el gráfico de la derecha
axes[1].set_title("Relación Y - X")
# Definimos la etiqueta del eje Y de la gráfica de la derecha
axes[1].set_ylabel("Eje Y")

figura.tight_layout()
```

### Tamaño de las figuras

```python
# Con figsize defino el tamaño
figura, axes = plt.subplots(1, 2, figsize = (10, 2))

axes[0].plot(x, y, color = "blue")
# Creamos un título para el gráfico de la izquierda
axes[0].set_title("Relación X - Y")
# Definimos la etiqueta del eje X de la gráfica de la izquierda
axes[0].set_xlabel("Eje X")

axes[1].plot(y, x, color = "red")
# Creamos un título para el gráfico de la derecha
axes[1].set_title("Relación Y - X")
# Definimos la etiqueta del eje Y de la gráfica de la derecha
axes[1].set_ylabel("Eje Y")

figura.tight_layout()
```

### Leyendas y labels

```python
figura, axes = plt.subplots(1, 2)
# Con label agrego una etiqueta; con $ obtengo una notación matemática
axes[0].plot(x, y, color = "blue", label = "$sin(x)$")
# Creamos un título para el gráfico de la izquierda
axes[0].set_title("Relación X - Y")
# Definimos la etiqueta del eje X de la gráfica de la izquierda
axes[0].set_xlabel("Eje X")
# Con legend muestro las etiquetas que había aplicado en la creación del plot
axes[0].legend()

axes[1].plot(y, x, color = "red")
# Creamos un título para el gráfico de la derecha
axes[1].set_title("Relación Y - X")
# Definimos la etiqueta del eje Y de la gráfica de la derecha
axes[1].set_ylabel("Eje Y")

figura.tight_layout()
```

Puedo modificar la localización de la leyenda:

- upper right.
- upper left.
- upper center.
- lower center.
- lower left.
- lower right.

```python
figura, axes = plt.subplots(1, 2)
# Con label agrego una etiqueta; con $ obtengo una notación matemática
axes[0].plot(x, y, color = "blue", label = "$sin(x)$")
# Creamos un título para el gráfico de la izquierda
axes[0].set_title("Relación X - Y")
# Definimos la etiqueta del eje X de la gráfica de la izquierda
axes[0].set_xlabel("Eje X")
# Con legend muestro las etiquetas que había aplicado en la creación del plot
axes[0].legend(loc = "upper center")

axes[1].plot(y, x, color = "red")
# Creamos un título para el gráfico de la derecha
axes[1].set_title("Relación Y - X")
# Definimos la etiqueta del eje Y de la gráfica de la derecha
axes[1].set_ylabel("Eje Y")

figura.tight_layout()
```

## Iterar a través de los axes

```python
# Definimos o pedimos el número de filas y columnas
filas = 3; columnas = 3

# Creamos la figura y las gráficas con el número de filas y columnas especificadas
figura, axes = plt.subplots(filas, columnas, figsize = (12, 6))

# Comenzamos a iterar por filas
for fila in range(filas):
    # Comenzamos a iterar por columnas
    for columna in range(columnas):
        # Creamos los axes o gráficos en la respectiva fila y columna
        axes[fila, columna].plot(x, np.sin(x), color = "red")
        axes[fila, columna].set_title("$sin(x)$")

# Por último, mostramos todos los gráficos
figura.tight_layout()
```

Personalizar los colores y estilos.

```python
figura, axes = plt.subplots(figsize = (6, 6))

# Color RGB
axes.plot(x, x + 1, color = "#D426C8")
axes.plot(x, x + 2, ":", color = "#66DF89")

# Colores comunes
axes.plot(x, x + 3, "-", color = "blue")
axes.plot(x, x + 4, "--", color = "red")
```

Con *alpha* podemos modificar la transparencia y con *linewidth* el grosor de la línea.

```python
figura, axes = plt.subplots(figsize = (6, 6))

# Modificando la transparencia y el grosor
axes.plot(x, x + 1, color = "#D426C8", alpha = 0.5, linewidth = 18)

# Modificando el grosor
axes.plot(x, x + 2, ":", color = "#66FF89", linewidth = 3)

# Modificando el grosor
axes.plot(x, x + 3, "-", color = "blue", linewidth = 5)

# Modificando la transparencia y el grosor
axes.plot(x, x + 4, "--", color = "red", alpha = 0.3, linewidth = 9)
```

Modificando el estilo de la línea y los marcadores con *linestyle* y *marker*.

```python
figura, axes = plt.subplots(figsize = (6, 6))

# Modificando el estilo de la línea
axes.plot(x, x + 1, color = "#D426C8", linestyle = "-.")

# Modificando el marcador
axes.plot(x, x + 2, color = "#66FF89", marker = "X")

# Modificando el estilo de la línea y el marcador
axes.plot(x, x + 3, color = "blue", linestyle = "--", marker = "v")

axes.plot(x, x + 4, color = "red")
```

Con *markersize* y *markerfacecolor* podemos modificar el tamaño y color del marcador.

```python
figura, axes = plt.subplots(figsize = (6, 6))

# Modificando el estilo de la línea 
axes.plot(x, x + 1, color = "#D426C8", linestyle = "-.")

# Modificando el marcador y su tamaño
axes.plot(x, x + 2, color = "#66FF89", marker = "X", markersize = 10)

# Modificando el estilo de la línea y el marcador junto a su tamaño y color
axes.plot(x, x + 3, color = "blue", linestyle = "--", marker = "v", markersize = 15, markerfacecolor = "#FF0000")

axes.plot(x, x + 4, color = "red")
```

## Barplot

El gráfico de barras nos permite graficar variables categóricas.

```python
countrys = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", 'GERMANY']
population = [2000, 800, 900, 1000, 300]
```

Gráfica de barras; primero las categorías y luego los valores.

```python
figura, axes = plt.subplots(figsize = (6, 6))
# (categorías, valores)
axes.bar(countrys, population)
figura.tight_layout()
```

Podemos modificar los colores y el ancho de las barras.

```python
figura, axes = plt.subplots(figsize = (6, 6))
# (categorías, valores)
axes.bar(countrys, population, width = 0.5, color = ["#264653", "#2a9d8f", "#e9c46a", "#f4a261", "#e76f51"])
figura.tight_layout()
```

También podemos modificar los nombres de las categorías con *xticks* y rotar dichas etiquetas con *rotation*.

```python
plt.bar(countrys, population, width = 0.5, color = ["#264653", "#2a9d8f", "#e9c46a", "#f4a261", "#e76f51"])
plt.xticks(np.arange(0, 5), ('India','Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.show()
```

Es posible realizar un barplot horizontal.

```python
plt.barh(countrys, population, height = 0.5, color = ["#264653", "#2a9d8f", "#e9c46a", "#f4a261", "#e76f51"])
plt.yticks(np.arange(0, 5), ('India','Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.show()
```

## Histogramas

```python
data = np.random.randint(1,100,50)
```

Crear un histograma

```python
plt.hist(data)
plt.show()
```

Con *bins* podemos indicar el número de barras y con *width* podemos cambiar el ancho de las barras.

```python
plt.hist(data, bins = 50, width = 0.5)
plt.show()
```

Podemos cambiar el estilo o tipo del histograma.

```python
plt.hist(data, histtype = "step", color = "red")
plt.show()
```

## Boxplot

Nos muestra el rango intercuartil y la mediana.

```python
plt.boxplot(data)  #Muestra rango intercuartil y la mediana
plt.show()
```

Con *vert* puedo cambiar la dirección, con *patch_artist* se rellena el rango intercuartil y con *notch* se enfoca la mediana.

```python
plt.boxplot(data, vert = False, patch_artist = True, notch = True)
plt.show()
```

Se pueden quitar los outliers con *showfliers*.

```python
data = np.append(data, 400)
plt.boxplot(data, vert = False, patch_artist = True, notch = True, showfliers = False)  
plt.show()
```

## Scatter plot

Gráfico de dispersión con área.

```python
plt.scatter(x, y, s = area)
plt.show()
```

Cambiar el color y cambiar los marcadores.

```python
plt.scatter(x, y, s = area, c = colors, marker = "X")
plt.show()
```

Agregar transparencia.

```python
plt.scatter(x, y, s = area, c = colors, marker = "o", alpha = 0.5)
plt.show()
```
